```python
#7
import pandas as pd


data = {'Phone': ['123-456-7890', '555-123-4567', 'abc-def-123', '9876543210', 'xxx-yyy-zzz']}
df = pd.DataFrame(data)


def clean_phone_number(phone_number):
    """
    Cleans a phone number by removing non-numeric characters.

    Args:
        phone_number: The phone number string to clean.

    Returns:
        The cleaned phone number as a string, or None if the phone number is empty.
    """
    if not phone_number:
        return None
    return ''.join(char for char in phone_number if char.isdigit())

df['Cleaned_Phone'] = df['Phone'].apply(clean_phone_number)


df['Phone_Number'] = pd.to_numeric(df['Cleaned_Phone'], errors='coerce')


print(df.info())
```

```python
#13
import numpy as np
import matplotlib.pyplot as plt

# Create two NumPy arrays with 100 random float values between 0 and 1
x = np.random.rand(100)
y = np.random.rand(100)

# Create a scatter plot
plt.scatter(x, y, color='red', marker='o', label='Data Points')

# Add horizontal and vertical lines
plt.axhline(y=0.5, color='green', linestyle='--', label='y = 0.5')
plt.axvline(x=0.5, color='blue', linestyle=':', label='x = 0.5')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Scatter Plot of Random Values')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

```python
#14
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
dates = pd.date_range(start='2023-01-01', end='2023-12-31')
temperature = np.random.randint(10, 40, size=len(dates))
humidity = np.random.randint(40, 100, size=len(dates))

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Set 'Date' as index
df.set_index('Date', inplace=True)

# Plotting
fig, ax1 = plt.subplots()

# Plot Temperature
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color=color)
ax1.plot(df.index, df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for Humidity
ax2 = ax1.twinx()

# Plot Humidity
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(df.index, df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Customize plot
plt.title('Time Series Data')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#16
import numpy as np
import matplotlib.pyplot as plt

# Create a NumPy array with 1000 samples from a standard normal distribution
data = np.random.randn(1000)

# Create a histogram
plt.hist(data, bins=50, density=True, alpha=0.7, color='blue', label='Histogram')

# Plot the PDF
x = np.linspace(-4, 4, 100)
pdf = 1 / np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add labels
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#18
from bokeh.plotting import figure, show
from bokeh.models import Grid

# Generate data for the sine wave
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create the figure object
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add the line to the plot
p.line(x, y, line_width=2)

# Add grid lines
grid = Grid()
p.add_layout(grid)

# Show the plot
show(p)
```

```python
#19
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, FactorRange
from bokeh.palettes import Category20c
import numpy as np

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
counts = np.random.randint(1, 20, size=len(categories))

# Create ColumnDataSource for efficient data handling
source = ColumnDataSource(data=dict(
    x=categories,
    top=counts,
    color=Category20c[len(categories)]
))

# Create the figure
p = figure(x_range=FactorRange(*categories),
           title="Random Categorical Bar Chart",
           toolbar_location=None,
           tools="")

# Create the bars
p.vbar(x='x', top='top', width=0.9, color='color', source=source)

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@x"), ("Count", "@top")])
p.add_tools(hover)

# Customize axes and labels
p.xgrid.grid_line_color = None
p.y_axis.axis_label = "Count"

# Show the plot
show(p)
```
